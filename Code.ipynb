## 13. Do students in leadership positions during their college years tend to have higher GPAs or better expected salary?

## 18. How many students know about the event from their colleges? Which are the Top 5 colleges?

```python
columns_to_process=['Source','Other_Source']
# Convert all values in Column2 to strings
df[columns_to_process] = df[columns_to_process].astype(str)
```

```python
# Function to replace "Others" within strings in Column1
def replace_others(row):
    if "Others" in row['Source']:
        return row['Source'].replace("Others", row['Other_Source'])
    return row['Source']

# Apply the function to replace "Others" in Column1
df['Source'] = df.apply(replace_others, axis=1)
df=df.drop('Other_Source',axis=1)
```

```python
# Define substrings to search for
substrings = ["College", "College Professor"]
# Create a regex pattern to match any of the substrings
pattern = '|'.join(substrings)
# Filter the DataFrame
filtered_df = df[df['Source'].str.contains(pattern, case=False, na=False)]
print("No of Students those who know about event through college =",len(filtered_df))
```

## No of Students those who know about event through college = 498

```python
# Get the top 5 college counts
top_colleges = filtered_df['College Name'].value_counts()[:5]
# Create a bar plot
top_colleges.plot(kind='bar', figsize=(6, 4),fontsize=7)
# Add labels and title
plt.title('Top 5 College who Informed Students about Event')
plt.xlabel('College Name')
plt.ylabel('Count')
plt.show()
```
